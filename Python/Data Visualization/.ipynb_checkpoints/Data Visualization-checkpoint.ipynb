# TESTING

```python
import pandas as pd
import numpy as np
import scipy.stats
from matplotlib import pyplot as plt
```

```python
x = range(0,10)
y = list(map(lambda x: x**2, x))
z = list(map(lambda x: x**2.1, x))
plt.plot(x,y)
plt.plot(x,z)
plt.title("Square Value")
plt.xlabel("x")
plt.ylabel("y and z")
plt.legend(["x^2","x^2.1"])
plt.show
```

# BENERAN

```python
info = pd.read_csv('00. InfoProduct.csv')
info.replace('Rp. 0', np.nan, inplace = True)
info.dropna(inplace = True)
info.reset_index(inplace = True)
info['Reviewer'] = info['Reviewer'].astype('int64')
info.head(5)
```

```python
Cost = []
for i in range(len(info.Price)):
    a = info.Price[i].split('.')
    b = ""
    for i in range(1,len(a)):
        b += a[i]
    Cost.append(int(b))
info["Cost"] = Cost
info.drop(columns = 'Price', inplace = True)
info
```

```python
dataDesc = []
for i in info.columns:
    dataDesc.append([
        i,
        info[i].dtypes,
        info[i].isnull().sum(),
        round((((info[i].isnull().sum())/(len(info)))*100),3),
        info[i].nunique(),
        info[i].drop_duplicates().sample(3).values
    ])
pd.DataFrame(dataDesc, columns = [
    'Data Features',
    'Data Types',
    'Null',
    'Null Percentage',
    'Unique',
    'Unique Sample'
])
```

```python
info.drop((info[info['Cost']>3000000]).index, inplace = True)
plt.scatter(info.Cost, info.OverallRating, color = 'b')
plt.title("Price vs. Rating")
plt.xlabel("Price (Rupiah)")
plt.ylabel("y-axis")
plt.grid()
```

```python
plt.scatter(info.Cost, info.Reviewer, color = 'b')
plt.title("Price vs. Reviewer")
plt.xlabel("Price (Rupiah)")
plt.ylabel("Number of Reviewer")
plt.grid()
```

```python
scipy.stats.pearsonr(info.Cost, info.OverallRating)
```

```python
from textwrap3 import wrap
cat = info.Category.unique()
category = [ '\n'.join(wrap(i, 5)) for i in category ]
counts = []
for i in cat:
    counts.append(info["Category"][info.Category == i].count())
fig = plt.figure(figsize=(25,10))
ax = plt.axes()
ax.bar(category, counts)
plt.show()
```

```python
fig = plt.figure(figsize=(10,6))
ax = plt.axes()
histx = [1,2,3,4,5]
ax.hist(info.OverallRating, histx, histtype = 'bar', rwidth = 0.9)
plt.show()
```

```python
summ = []
for i in range(len(cat)):
    summ.append(info.Cost[info.Category == cat [i]].sum())
    
def absolute_value(x):
    a = np.round(x/100.*sum(Cost), 0)
    return a

fig = plt.figure(figsize = (15,15))
ax = plt.axes()
ax.pie(summ, labels = cat, startangle = 100, autopct = absolute_value)
plt.title("Sum of Revenue per Product Category")
plt.show()
```
